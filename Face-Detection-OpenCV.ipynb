```python
#import libraries
import cv2
import numpy as np 
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#create classifier
face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
#open image
image = cv2.imread('Desktop/download.jpeg')
fix_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
#Detect faces  
faces = face_classifier.detectMultiScale(image, 1.3, 5)
```

```python
#fuction for bounding box around the faces 
def detect_faces(fix_img):
    face_rects = face_classifier.detectMultiScale(fix_img, 1.005, 3)
    
    for (x,y,w,h) in face_rects:
        cv2.rectangle(fix_img,
                     (x,y),
                     (x+w, y+h),
                     (255,0,0),
                     10)
        return fix_img
    
```

```python
result = detect_faces(fix_img)
plt.imshow(result)
```
